```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
birds = pd.read_csv('birds.csv')
birds.head()
```

```python
birds.info()
```

```python
birds.describe()
```

# Line Chart

```python
snow_goose_df = birds[birds['Name'] == 'Snow goose']

plt.figure(figsize=(10, 5))
plt.plot(snow_goose_df['MinLength'], label='Min Length')
plt.plot(snow_goose_df['MaxLength'], label='Max Length')
plt.xlabel('Snow Goose Index')
plt.ylabel('Length (cm)')
plt.title('Minimum and Maximum Length of Snow Geese')
plt.legend()
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(12, 6))
plt.scatter(snow_goose_df['MinBodyMass'], snow_goose_df['MaxBodyMass'])
plt.xlabel('Min Body Mass (g)')
plt.ylabel('Max Body Mass (g)')
plt.title('Body Mass Range of Snow Geese')
plt.show()
```

# Bar Chart

```python
plt.figure(figsize=(10, 6))
plt.bar(snow_goose_df['ConservationStatus'], height=0.5)
plt.xlabel('Conservation Status')
plt.ylabel('Count')
plt.title('Conservation Status of Snow Geese')
plt.show()
```
